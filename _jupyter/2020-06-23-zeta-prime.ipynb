I came across an interesting [blog article](http://chrishenson.net/website/zeta.html) by Chris Henson on the relationship between the Riemann Zeta function and prime numbers. After the thoughtful read, I thought I'd write an article on this topic, as it is closely related to some of the other 

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from functools import reduce
from scipy.special import zeta
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
plt.style.use("seaborn")
```

```python
def is_coprime(n_num, max_int):
    '''
    Generate `n_nums` randon numbers
    within range 1 to `max_int` and
    check if they are coprime
    '''
    nums = np.random.randint(1, max_int, size=n_num)
    return reduce(math.gcd, nums) == 1
```

```python
def test(n_num, max_int, n_iter):
    '''
    Return `x` `max_int` domain and
    `y`, a list of mean values of
    coprime test for given `max_int`
    '''
    x = range(2, max_int, 1)
    y = [np.mean([is_coprime(n_num, max_int) for _ in range(n_iter)]) for max_int in x] 
    return x, y
```

```python
x, y = test(2, 200, 1000)
```

```python
plt.scatter(x, y, color="skyblue")
plt.hlines(y=1/zeta(2), xmin=2, xmax=200, 
           colors="gold", label = r'$\frac{1}{\zeta(2)}$')
plt.legend()
plt.plot()
```

```python
def test_n(n_nums):
    for n_num in n_nums:
        x, y = test(n_num, 200, 1000)
        plt.scatter(x, y)
        plt.hlines(y=1/zeta(n_num), xmin=2, xmax=200, 
           colors="gold")
    plt.legend()
```

```python
test_n(range(2, 5, 1))
```
